```python
x = torch.tensor(2.0, requires_grad=True)
y = x ** 2 + 3*x + 1

y.backward()
print(x.grad)
```

# y = wx + b

```python

w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)

x = torch.tensor(2.0)
y = w*x + b

y.backward()

print(w.grad, b.grad)
```

# Using torch.nn Module

```python
import torch.nn as nn

model = nn.Linear(1, 1)
print(model)
```

# Forward Pass

```python
x = torch.tensor([[2.0]])
y_pred = model(x)

print(y_pred)
```

# Loss Function

```python
criterion = nn.MSELoss()

y_true = torch.tensor([[5.0]])
loss = criterion(y_pred, y_true)

print(loss)
```

# Optimizer

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

# Training Step

```python
optimizer.zero_grad()
loss.backward()
optimizer.step()

print("Updated weights:", model.weight.data, model.bias.data)
```

# Mini Training Loop

```python
for epoch in range(5):
    optimizer.zero_grad()
    y_pred = model(x)
    loss = criterion(y_pred, y_true)
    loss.backward()
    optimizer.step()
    
    print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```
